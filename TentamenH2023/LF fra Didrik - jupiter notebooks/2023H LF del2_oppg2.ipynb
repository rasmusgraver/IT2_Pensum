```python
# Importerer nødvendige bibliotek
import csv
import matplotlib.pyplot as plt
```

```python
# Henter data fra filen
filnavn = 'transistorer.csv'
tid = []
transistorer = []
with open(filnavn, 'r', encoding='utf-8-sig') as fil:
    innhold = csv.reader(fil, delimiter=',')
    overskrifter = next(innhold)

    for rad in innhold:
        # Gjør om til integer, siden både årstall og antall transistorer er et heltall 
        tid.append(int(rad[0]))
        transistorer.append(int(rad[1]))
```

```python
plt.title("Foventet antall transistorer i en CPU")
plt.plot(tid, transistorer, 'o')
plt.xlabel('Årstall')
plt.ylabel('Antall transistorer')
plt.yscale('log')

plt.show()
```

```python
moores_lov = [2300]
aarstall = [1971]

# Dobler antall transistorer hvert andre år fram til 2023
while aarstall[-1] < 2023:
    aarstall.append(aarstall[-1] + 2)
    moores_lov.append(moores_lov[-1] * 2)
```

```python
plt.subplot(2, 1, 1)
plt.title("Uten logaritmisk skala")
plt.plot(aarstall, moores_lov, 'r', label="Moores lov")
plt.plot(tid, transistorer, 'x', label="Data")
plt.xlabel('Årstall')
plt.ylabel('Antall transistorer')
#plt.yscale('log')
plt.legend()

plt.subplot(2,1,2)

plt.title("Med logaritmisk skala")
plt.plot(aarstall, moores_lov, 'r', label="Moores lov")
plt.plot(tid, transistorer, 'x', label="Data")
plt.xlabel('Årstall')
plt.ylabel('Antall transistorer')
plt.yscale('log')
plt.legend()

plt.tight_layout()

plt.show()
```

Fra grafen uten logaritmisk y-akse ser det ut som Moores lov ikke lenger er gyldig; "loven" overestimerer antall transistorer siden 2020. Likevel ser vi fra grafen med logaritmisk y-akse at det relative avviket ikke er større nå enn perioden 1990-2000. 

Vi kan dermed si at Moores lov fremdeles er en god modell, men da den konsekvent har overestimert antall transistorer siden ca. 2010 forventes det nok ikke at dette vedvarer veldig mange år fremover (esponentiell vekst må før eller siden stoppe).

```python
# Bruker Moores lov fram til 2031
while aarstall[-1] < 2031:
    aarstall.append(aarstall[-1] + 2)
    moores_lov.append(moores_lov[-1] * 2)

maal = [2025, 2031]

for y in maal:
    i = aarstall.index(y)
    # Gjør utskriften på standardform slik at den er enklere å lese
    print(f"I {y} er et forventet antall transistorer ifølge Moores lov {moores_lov[i]:.3e}.")
```
